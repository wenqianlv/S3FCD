## 特征图

```python
import random
# random.seed(318)
fs1 = feat1
fs2 = feat2
nr,nc = 10,6
fig, ax = plt.subplots(nrows=nr, ncols=nc, figsize=(nc*2, nr*2), squeeze=False)
camp = 'inferno'
for i in range(0,nr,2):
    if len(fs1) <= i//2: break
    for j in range(nc):
        if len(fs1[i//2]) <= j: break
        jdx = random.randint(0,len(fs1[i//2])-1)
        ax[i][j].imshow(fs1[i//2][jdx], cmap=camp)
        ax[i+1][j].imshow(fs2[i//2][jdx], cmap=camp)
plt.show()
```
